b. bina dataframe

c. gunakan data markah tadi untuk menghasilkan bar chart/ graf perlu menunjukkan nama pelajar di paksi X dan Markah di Paksi Y

```python
# Import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
nama_pelajar = ["Ali", "Abu", "Siti", "Aina", "John"]
markah = np.array([80, 75, 90, 85, 70])

# (a) Analisis Menggunakan NumPy
print("Semua markah :", markah)
print("Markah tertinggi :", np.max(markah))
print("Markah terendah :", np.min(markah))
print("Purata markah :", np.mean(markah))
print("Jumlah markah :", np.sum(markah))

# (b) Bina DataFrame
df = pd.DataFrame({
    "Nama Pelajar": nama_pelajar,
    "Markah": markah
})

print("\nDataFrame")
print(df)

# (c) Bar Chart Menggunakan Matplotlib
plt.figure(figsize=(6,4))
plt.bar(df["Nama Pelajar"], df["Markah"])

plt.title("Prestasi Markah Pelajar")
plt.xlabel("Nama Pelajar")
plt.ylabel("Markah")
plt.ylim(0, 100)

plt.show()
```
